```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("Telco-Customer-Churn.xlsx")
```

```python
df
```

# 1. Compare churn count with respect to gender.

```python
df["gender"].value_counts()
```

```python
df["Churn"].value_counts()
```

```python
sns.countplot(x="gender", data = df, hue="Churn")
plt.title("Churn vs Gender")
plt.xlabel("Gender")
plt.ylabel("Churn count")
plt.show()
```

# 2. Find out how many female senior citizens there in the dataset

```python
df1= df[(df["gender"]=="Female") & (df["SeniorCitizen"]==1)]
```

```python
df1
```

# 3. Compare 'tenure' with 'Total Charges'

```python
plt.figure(figsize=(9,10))
plt.scatter(df['tenure'],df['TotalCharges'],color="red")
plt.title("scatter plot of Tenure and Total Charges")
plt.xlabel("Tenure")
plt.ylabel("TotalCharges")
```

# 4. Find out which contract is preferred by the senior citizen.

```python
plt.figure(figsize=(9,10))
sns.countplot(df["SeniorCitizen"], hue =df["Contract"])
plt.title("Contract preferred by the senior citizen")
plt.xlabel("Senior citizen")
plt.ylabel("count of contract")
plt.show()
```

# 5. Comment on your finds on Payment Method
